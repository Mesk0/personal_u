```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(u):
    return u**3 - 5*u**2 + 9

def exp(u):
    return np.e**u

x = np.linspace(-5, 5, num=100)
y = f(x)
plt.plot(x, y)
plt.plot(x, exp(x))
```

```python
def exp2(x):
    sum = 0
    for k in range(100):
        sum += float(x**k)/np.math.factorial(k)
    return sum

print(exp2(2), exp2(3))
```
